```python
import spacy 
import nltk # natural language package similar to spacy 
import string
import re # regular expression
from gensim.models import Word2Vec
```

```python
# download punct - word tokenization , stopword- end word identify
nltk.download('punkt')
nltk.download('stopwords')
```

```python
#data reading
train_med = "train.dat"   # training data taken 
v = open(train_med, 'r')  # r - read data 
text = v.read()
```

```python
# text cleaning / sentence segementation / removal of punct
text_clean = "".join([i for i in text if i not in string.punctuation])

```

```python
print(text_clean[0:40])
```

```python
# convert to lower case
text_lowered = "".join([i.lower() for i in text_clean])
```

```python
print(text_lowered[0:90])
```

```python
#tokenization 
words = nltk.tokenize.word_tokenize(text_lowered)
```

```python
print(words[0:40])
```

```python
stopwords = nltk.corpus.stopwords.words('english')
#initialise stop words
```

```python
print(stopwords[0:40])
```

```python
words_new = [i for i in words if i not in stopwords]
```

```python
words_new
```

```python
#stemming
ss = nltk.SnowballStemmer(language = 'english')
```

```python
w = [ss.stem(word) for word in words_new]
```

```python
w[0:100]
```

```python
model_med = Word2Vec(
    sentences=[words_new],
    vector_size=3600,
    workers=4,
    window = 15,
    epochs= 15,
    sg=1)
```

```python
wordvect = model_med.wv
```

```python
len(wordvect.index_to_key)
```

```python
wordvect.most_similar('angioplasty',topn=20)
```

```python
#lemmatization
from nltk.stem import WordNetLemmatizer
wordnet_lemmatizer = WordNetLemmatizer()
```

```python
nltk.download('wordnet')
nltk.download('omw-1.4')
```

```python
w_lem = [wordnet_lemmatizer.lemmatize(word) for word in words_new]
```

```python
model_med_lm = Word2Vec(
    sentences=[w_lem],
    vector_size=3600,
    workers=4,
    window = 15,
    sg=1)
```

```python
wordvectlm = model_med_lm.wv
wordvectlm.most_similar('cardiopulmonary', topn=20)
```
